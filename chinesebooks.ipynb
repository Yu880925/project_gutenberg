```python
from bs4 import BeautifulSoup as bs
from pprint import pprint
from selenium import webdriver
from time import sleep
import requests as req
import json, re, os

```

```python
os.makedirs('project_gutenberg', exist_ok=True)



url = 'https://www.gutenberg.org/browse/languages/zh'
res = req.get(url)
soup = bs(res.text, 'lxml')

#建立list放資訊
list_posts = []
list_posts.clear()

#開啟網頁
driver = webdriver.Chrome()

#取得中文書的書名與連結
for a in soup.select('li.pgdbetext a'):
    href = a['href']
    title = a.get_text().strip()
    if re.match(r'^[\u4E00-\u9FFF]+$', title):
        new_ = title
        full_url = 'https://www.gutenberg.org' + href

        
        list_posts.append({
            'title': new_,
            'link': full_url
        })


for index, obj in enumerate(list_posts):
    try:
        res_ = req.get(obj['link'])
        soup_ = bs(res_.text, "lxml")
        txt_link = soup_.find('a', string = 'Plain Text UTF-8')    #找出連結名為Plain Text UTF-8

        if not txt_link:
            continue

        txt_href = txt_link['href']
        if not txt_href.startswith('http'):
            txt_url = 'https://www.gutenberg.org' + txt_href
        else:
            txt_url = txt_href

        #進到中文書內容txt  
        driver.get(txt_url)

        res_txt = req.get(txt_url)
        res_txt.encoding = 'utf-8'
        text_ = res_txt.text

        # 找出中文和標點
        chinese_parts = re.findall(r'[\u4E00-\u9FFF\u3000-\u303F，。！？：「」、．《》…—\[\];*]+', text_)
        chinese_text = '\n'.join(chinese_parts)


        #建立中文書的txt檔
        with open(f'project_gutenberg/{obj["title"]}.txt', 'w', encoding='utf-8') as f:
            f.write(chinese_text)
        
        print(f'{obj[title]}-下載完成')
    
    except Exception:
        print("錯誤")
    


print(list_posts)
print(len(list_posts))
```
